```python
import dask
import dask.dataframe as dd
import pandas as pd
from datetime import datetime
```

```python
parse_dates = ['departure', 'return']
```

```python
df = dd.read_csv('database.csv', dtype={'departure': 'string',
                                        'return': 'string',
                                        'departure_name':'string',
                                        'return_name':'string',
                                        'departure_id': 'float64',
                                         'return_id': 'float64',
                                        'distance(m)':'float64',
                                        'duration(sec.)':'float64',
                                        'avg_speed (km/h)': 'float64',
                                        'departure_latitude':'float64',
                                        'departure_longitude':'float64',
                                        'return_latitude':'float64',
                                        'return_longitude':'float64',
                                        'Air temperature (degC)':'float64'
                                       }, parse_dates = parse_dates)
df
```

```python
def to_months(x):
    return x.strftime('%Y-%m')
```

```python
df = df.assign(months = df["departure"].map(to_months))
```

```python
df
```

```python
df.repartitionByRange('months').write.csv("data/months")
```

```python
df.divisions = tuple(df['months'])
```
